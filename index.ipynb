```python
df = pd.DataFrame(columns = ['title', 'price','link_item'])
for i in tqdm(range(2, 40)):
    url = f'https://www.lazada.vn/catalog/?_keyori=ss&from=search_history&page={i}&q=dien%20thoai%20di%20dong&spm=a2o4n.10441748.search.2.69ef4249fsMW4A&sugg=dien%20thoai%20di%20dong_0_1'
    driver.get(url)
    elems = driver.find_elements(By.CSS_SELECTOR, '.RfADt [href]')
    title = [elem.text for elem in elems]
    links = [elem.get_attribute('href') for elem in elems]
    elems_price = driver.find_elements(By.CSS_SELECTOR, ".aBrP0")
    price = [elem_price.text for elem_price in elems_price]
    elems_countReviews = driver.find_elements(By.CSS_SELECTOR,'._6u7NR')
    countReviews = [elem.text for elem in elems]
    elems_location = driver.find_elements(By.CSS_SELECTOR,'.oa6ri')
    locations = [elem.text for elem in elems_location]
    elems_discount = driver.find_elements(By.CSS_SELECTOR,'.IcOsH')
    discount = [elem.text for elem in elems_discount]
    for title, price, link,review,location in zip(title, price, links,countReviews,locations):
        df = df.append({'title': title, 'price': price, 'link_item': link, 'Review':review}, ignore_index=True)
```
